## Read the CSV and Perform Basic Data Cleaning

```python
# Load the data
file_path = Path('Resources/LoanStats_2019Q1.csv')
df = pd.read_csv(file_path)

# Preview the data
df.head()
```

## Split the Data into Training and Testing

```python
pd.options.display.max_rows=100
columns=df.dtypes=='object'
list(columns[columns].index)

list_columns=list(columns[columns].index)
list_columns
```

```python
for i in list_columns:
    le = LabelEncoder()

# Encoding homeowner column
    le.fit(df[i])
    df[i] = le.transform(df[i])
```

```python
df.head()
```

```python
le = LabelEncoder()

# Encoding homeowner column
le.fit(df["verification_status"])
df["verification_status"] = le.transform(df["verification_status"])

df.head()
```

```python
# Create our features
X= df.drop(columns="loan_status")

# Create our target
y=df["loan_status"]
```

```python
X.describe()
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
# Split the X and y into X_train, X_test, y_train, y_test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test= train_test_split(X,y, random_state=1,stratify=y)
```

## Data Pre-Processing

Scale the training and testing data using the `StandardScaler` from `sklearn`. Remember that when scaling the data, you only scale the features data (`X_train` and `X_testing`).

```python
# Create the StandardScaler instance
scaler = StandardScaler()
```

```python
# Fit the Standard Scaler with the training data
# When fitting scaling functions, only train on the training dataset
X_scaler = scaler.fit(X_train)
```

```python
# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Balanced Random Forest Classifier

```python
# Resample the training data with the BalancedRandomForestClassifier
from imblearn.ensemble import BalancedRandomForestClassifier

balanced_random_forest = BalancedRandomForestClassifier(n_estimators=1000, random_state=1)
balanced_random_forest.fit(X_train_scaled, y_train)
```

```python
# Calculated the balanced accuracy score
y_pred_brf = balanced_random_forest.predict(X_test_scaled)
balanced_as_brf=balanced_accuracy_score(y_test, y_pred_brf)
print(balanced_as_brf)
```

```python
# Display the confusion matrix
cm_brf = confusion_matrix(y_test, y_pred_brf)
cm_df_brf = pd.DataFrame(
    cm_brf, index=["Actual High Risk", "Actual Low Risk"], columns=["Predicted High Risk", "Predicted Low Risk"]
)
cm_df_brf
```

```python
# Print the imbalanced classification report
print(classification_report_imbalanced(y_test, y_pred_brf))
```

```python
# List the features sorted in descending order by feature importance
importances = balanced_random_forest.feature_importances_
sorted(zip(balanced_random_forest.feature_importances_, X.columns), reverse=True)
```
